## 1. 取出今天所有的新聞

```python
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
from datetime import datetime, timedelta
today = datetime.now().strftime("%Y/%m/%d")
today
```

```python
import time
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    
    date = soup.find(class_="part_list_2").find_all("h3")[-1].find(class_="date").text
    date = datetime.strptime(date,"%Y/%m/%d %H:%M").strftime("%Y/%m/%d")
    if date!=today:
        print('done.')
        break
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
for d in soup.find(class_="part_list_2").find_all("h3"):
    date = d.find(class_="date").text
    date = datetime.strptime(date,"%Y/%m/%d %H:%M").strftime("%Y/%m/%d")
    if date < today:
        pass
    else:
        print(d.text)
```

## 2. 取出現在時間兩小時內的新聞

```python
two_hours_ago = datetime.now() - timedelta(hours=2)
two_hours_ago
```

```python
for d in soup.find(class_="part_list_2").find_all("h3"):
    date = d.find(class_="date").text
    date = datetime.strptime(date,"%Y/%m/%d %H:%M")
    if date < two_hours_ago:
        pass
    else:
        print(d.text)
```

```python
browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
today = datetime.now() 
three_days_ago = today - timedelta(3)

y, m, d  = three_days_ago.year, three_days_ago.month, three_days_ago.day
y, m, d
```

```python
three_days_ago = three_days_ago.strftime("%Y/%m/%d")

three_days_ago_3pm = three_days_ago + ' 15:00'
three_days_ago_5pm = three_days_ago + ' 17:00'

three_days_ago_3pm, three_days_ago_5pm
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
from selenium.webdriver.support.ui import Select

selectY = Select(browser.find_element_by_id("selY"))
selectY.select_by_value(str(y))
selectM = Select(browser.find_element_by_id("selM"))
selectM.select_by_value(str(m))
selectD = Select(browser.find_element_by_id("selD"))
selectD.select_by_value(str(d))
browser.find_element_by_id('button').click()
```

```python
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    
    date = soup.find(class_="part_list_2").find_all("h3")[-1].find(class_="date").text
    date = datetime.strptime(date,"%Y/%m/%d %H:%M")
    print(date)
    if date < datetime.strptime(three_days_ago_3pm,"%Y/%m/%d %H:%M"):
        break
```
